```python
# Task 1 - Add the sample_customer_data.csv into your sample_data folder

# Task 2 - Import all of the libraries that you will need
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Task 3 - load the data into a pandas dataframe
data = pd.read_csv('sample_customer_data.csv')

# Task 4 - describe the data - Is there anything that you notice with the number of rows for each column and the max value
print(data.describe())

# Task 5 - print out the first 5 rows of the data.  Do you get any more information ?
print(data.head())

# Task 6 - print out all the information of the data. We should now understand the overview of our data now. What is the information in each column, count and datatype
print(data.info())

# Task 7 - Print out the unique values for all the categorical data and the purchased column. Is there anything that you notice ?
print(data['gender'].unique())
print(data['city'].unique())
print(data['purchased'].unique())

# Task 8 - Print out any missing values
print("Missing values:\n", data.isnull().sum(), "\n")

# Task 9 - Think about what we need to do to this data :
# Remove missing values from age, income and gender
# Change the gender and city from categorical variables
# Find the outliers
# Add scaling to the columns

# Task 10 - complete all of the tasks from task 9

# Drop missing values
data = data.dropna()
print("Missing values:\n", data.isnull().sum(), "\n")

# Examine outliers with boxplots
# Age boxplot
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].boxplot(data['age'], vert=False)
axs[0].set_title('Outliers in Age')
axs[0].set_xlabel('Age')

# Income boxplot
axs[1].boxplot(data['income'], vert=False)
axs[1].set_title('Outliers in Income')
axs[1].set_xlabel('Income')
plt.tight_layout()
plt.show()

# Remove age outliers
ageQ1 = data['age'].quantile(0.25)
ageQ3 = data['age'].quantile(0.75)
ageIQR = ageQ3 - ageQ1
ageLowerBound = ageQ1 - 1.5 * ageIQR
ageUpperBound = ageQ3 + 1.5 * ageIQR
data = data[(data['age'] >= ageLowerBound) & (data['age'] <= ageUpperBound)]

# Remove income outliers
incomeQ1 = data['income'].quantile(0.25)
incomeQ3 = data['income'].quantile(0.75)
incomeIQR = incomeQ3 - incomeQ1
incomeLowerBound = incomeQ1 - 1.5 * incomeIQR
incomeUpperBound = incomeQ3 + 1.5 * incomeIQR
data = data[(data['income'] >= incomeLowerBound) & (data['income'] <= incomeUpperBound)]

# Remove income outlier
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].boxplot(data['age'], vert=False)
axs[0].set_title('Outliers in Age')
axs[0].set_xlabel('Age')

# Income boxplot
axs[1].boxplot(data['income'], vert=False)
axs[1].set_title('Outliers in Income')
axs[1].set_xlabel('Income')
plt.tight_layout()
plt.show()

print(data['gender'].unique())
print(data['city'].unique())

# Encode the data
encoded_data = pd.get_dummies(data[['gender', 'city']])
data = data.drop(['city', 'gender'], axis=1)
data = pd.concat([data, encoded_data], axis=1)
print(data.head())

# Add scaling to the columns
scaled_data1 = data[['age', 'income']]
data = data.drop(['age', 'income'], axis=1)
scaler = StandardScaler()
scaled_numeric = scaler.fit_transform(scaled_data1)
scaled_df = pd.DataFrame(scaled_numeric)
data = pd.concat([data, scaled_df], axis=1)
data.columns = list(data.columns[:-2]) + ['age', 'income']
print(data.head())
print(data.describe())

```
